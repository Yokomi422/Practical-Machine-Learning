```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid", context="talk", palette="Set2")
```

- sns.setというのは、matplotlibとseabornのデザインを変更するメソッドです。デザインは可視化の際に重要な要素ですが、デフォルトのままでも問題ありません。

## ヒストグラム

```python
# データの読み込み
df = sns.load_dataset("iris")
df.head()
```

```python
sns.histplot(df["sepal_width"], kde=False, bins=7)
```

#### スタージェスの公式

```python
np.log(2 * len(df)) + 1
```

### 複数のヒストグラムを描写

```python
data_setosa = df[df["species"] == "setosa"]
data_virginica = df[df["species"] == "virginica"]
```

```python
sns.histplot(data_setosa["sepal_width"], kde=False, bins=7, color="blue",alpha=0.5)
sns.histplot(data_virginica["sepal_width"], kde=False, bins=7, color="red",alpha=0.5)
```

### カテゴリごとにヒストグラムを描写

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="sepal_width",kde=True,bins=7, hue="species", multiple="dodge", shrink=0.8)
```

## 散布図

```python
# データの読み込み
df = sns.load_dataset("iris")
df.head()
```

```python
df["species"].unique()
```

```python
df.shape
```

- irisデータセットは、setosa・versicolor・virginicaの3種類のアヤメのがく片と花弁の長さと幅の計測値を集めたものです。

```python
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
```

### すべての数値データを散布図で可視化

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', size='petal_length')
```

### すべての数値データを散布図で可視化

```python
sns.pairplot(df, hue="species")
```
